```python
# นำเข้าไลบรารี pandas สำหรับจัดการข้อมูลตาราง
import pandas as pd

# อ่านไฟล์ CSV เข้ามาเก็บในตัวแปร df
df_rotten = pd.read_csv("rotten_tomatoes_critic_reviews.csv")

# แสดงข้อมูล 5 แถวแรก เพื่อตรวจสอบว่าโหลดสำเร็จหรือไม่
df_rotten.head()
```

```python
# ดูโครงสร้างข้อมูล
df_rotten.info()

# ดูชื่อคอลัมน์ทั้งหมด
print(df_rotten.columns)

# ตรวจสอบ missing value
df_rotten.isnull().sum()
```

```python
for col in df_rotten.columns:
    print(f"'{col}'")
```

```python
df_rotten.columns = df_rotten.columns.str.strip()
```

```python
df_rotten = df_rotten[['review_score', 'review_type']]
df_rotten.head()
```

```python
df_rotten = df_rotten.dropna()
```

```python
import re

# ลบ HTML tag
def remove_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', str(text))

df_rotten['review_score', 'review_type'] = df_rotten['review_score', 'review_type'].apply(remove_html)

# แปลงเป็น lowercase
df_rotten['review_score', 'review_type'] = df_rotten['review_score', 'review_type'].str.lower()

# ลบ punctuation
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df_rotten['review_score', 'review_type'] = df_rotten['review_score', 'review_type'].apply(remove_punctuation)
```
